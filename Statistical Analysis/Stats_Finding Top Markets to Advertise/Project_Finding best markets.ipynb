### Checking for Sample Representativeness

```python
#  Splitting each string in the column

interest_no_nulls =  df_coder_survey['JobRoleInterest'].dropna()
splitted_interests = interest_no_nulls.str.split(',')

#Frequency table describing the number of options

n_of_options = splitted_interests.apply(lambda x  : len(x))
n_of_options.value_counts(normalize = True).sort_index()*100
```

We can observe that only 31.7% of the respondents have given a single area of interest, while the others have multiple areas of interest. But since multiple courses are offered by the platform, this may actually be beneficial. 

Since our focus in web and mobile development, we will check how many respondents chose at least one of these two options. 

```python
import re

our_options = ['Web Developer', 'Mobile Developer']

def web_mobile_check(response):
    match_count = 0
    for option in our_options:
        match = re.search(option, response)
        if match:
            match_count +=1
            
    if match_count == 1 or match_count == 2:
        return True
    else:
        return False
    

web_or_mobile_dev = interest_no_nulls.apply(web_mobile_check)
```

```python
web_or_mobile_dev.value_counts(normalize = True)*100
```

```python
web_or_mobile = interest_no_nulls.str.contains(
    'Web Developer|Mobile Developer') # returns an array of booleans
freq_table = web_or_mobile.value_counts(normalize = True) * 100
print(freq_table)

print('\n')

%matplotlib inline
plt.style.use('fivethirtyeight')

freq_table.plot.bar()
plt.title('Percentage respondents interested in Web/Mobile Dev', y = 1.08)
plt.xticks( [0,1] , ['Web / Mobile Dev','Other'], rotation = 0,)
#plt.legend()
plt.ylim([0,100])
plt.show()
```

#### Thus we can conclude that most people (86%) in the survey are interested in Web or Mobile Development. Hence this sample can be considered representative of the population of interest. We need to determine further details of these respondents such as :

### Determining Paying Customers

#### We will analyze the 'MoneyForLearning' column to determine where the customers are willing to pay. 

The MoneyForLearning column describes in American dollars the amount of money spent by participants from the moment they started coding until the moment they completed the survey. Our company sells subscriptions at a price of $59 per month, and for this reason we're interested in finding out how much money each student spends per month.

We'll narrow down our analysis to only four countries: the US, India, the United Kingdom, and Canada. We do this for two reasons:

These are the countries having the highest frequency in the frequency table above, which means we have a decent amount of data for each.
Our courses are written in English, and English is an official language in all these four countries. The more people know English, the better our chances to target the right people with our ads.
Let's start with creating a new column that describes the amount of money a student has spent per month so far. To do that, we'll need to divide the MoneyForLearning column to the MonthsProgramming column. The problem is that some students answered that they have been learning to code for 0 months (it might be that they have just started). To avoid dividing by 0, we'll replace 0 with 1 in the MonthsProgramming column.

```python
# Replace 0s with 1s to avoid division by 0
fcc_pop['MonthsProgramming'].replace(0,1, inplace = True)

# New column for the amount of money each student spends each month
fcc_pop['money_per_month'] = fcc_pop['MoneyForLearning'] / fcc_pop['MonthsProgramming']
fcc_pop['money_per_month'].isnull().sum()
```

#### Keeping only those values where the money per month is not null.

```python
fcc_pop = fcc_pop[fcc_pop['money_per_month'].notnull()]
```

```python
fcc_money_per_month = fcc_pop.groupby('CountryLive').mean()['money_per_month']
```

```python
freq_coder_countries.head(4).index
```

```python
fcc_money_per_month[freq_coder_countries.head(4).index]
```

#### Thus we can observe that in USA and India, people are the highest spenders on a per month basis. It is surprising to note that UK and Canada spend less than India in spite of the higher per capita incomes. We will try to analyze the outliers in the next step. 

### Dealing with Outliers

We will create box plots to observe the outliers

```python
top_countries_bool = (fcc_pop['CountryLive'] == 'India') | (fcc_pop['CountryLive'] == 'United Kingdom') | (fcc_pop['CountryLive'] == 'Canada')|(fcc_pop['CountryLive'] == 'United States of America')

fcc_top_countries = fcc_pop[top_countries_bool]
```

```python
import seaborn as sns

sns.boxplot(x = 'CountryLive', y = 'money_per_month', data = fcc_top_countries )
plt.xlabel('Country of Residence')
plt.ylabel('Money Per Month (USD)')
plt.title('Monthly Amount Spent on Learning', y = 1.02)
plt.xticks([0,1,2,3], ['US', 'UK', 'India','Canada'], rotation = 0)
plt.show()
```

#### Dropping the Outlier values (any spends more than 20000 USD)

```python
fcc_top_countries = fcc_top_countries[fcc_top_countries['money_per_month'] < 20000]
```

```python
sns.boxplot(x = 'CountryLive', y = 'money_per_month', data = fcc_top_countries )
plt.xlabel('Country of Residence')
plt.ylabel('Money Per Month (USD)')
plt.title('Monthly Amount Spent on Learning', y = 1.02)
plt.xticks([0,1,2,3], ['US', 'UK', 'India','Canada'], rotation = 0)
#plt.ylim([0,80000])
plt.show()
```

```python
fcc_money_per_month_updated = fcc_top_countries.groupby('CountryLive').mean()['money_per_month']
```

```python
fcc_money_per_month_updated.sort_values(ascending = False)
```

#### We can still observe some outliers for India and the US. We will examine them one by one. 

```python
bool_india_outliers = (fcc_top_countries['CountryLive'] == 'India') & (fcc_top_countries['money_per_month'] > 2500)
india_outliers = fcc_top_countries[bool_india_outliers]
india_outliers
```

#### Since none of these coders attended a bootcamp (which could have inflated the spend), it is possible that these respondents included their University Tuition costs. Our objective is to figure out the money spent outside of the College / University system. Though we cannot conclusively attribute these values, it would be safe to remove these six values.

```python
fcc_top_countries.shape
```

```python
fcc_top_countries = fcc_top_countries.drop(india_outliers.index)
fcc_top_countries.shape
```

```python
bool_us_outliers = (fcc_top_countries['CountryLive'] == 'United States of America') & (fcc_top_countries['money_per_month'] > 6000)
us_outliers = fcc_top_countries[bool_us_outliers]
us_outliers
```

Out of these 11 extreme outliers, six people attended bootcamps, which justify the large sums of money spent on learning. For the other five, it's hard to figure out from the data where they could have spent that much money on learning. Consequently, we'll remove those rows where participants reported thay they spend $6000 each month, but they have never attended a bootcamp.

Also, the data shows that eight respondents had been programming for no more than three months when they completed the survey. They most likely paid a large sum of money for a bootcamp that was going to last for several months, so the amount of money spent per month is unrealistic and should be significantly lower (because they probably didn't spend anything for the next couple of months after the survey). As a consequence, we'll remove every these eight outliers.

In the next code block, we'll remove respondents that:

1. Didn't attend bootcamps.
2. Had been programming for three months or less when at the time they completed the survey.

```python
## Removing the participants that didn't attend a bootcamp

bool_no_bootcamp = bool_us_outliers & (fcc_top_countries['AttendedBootcamp'] == 0)
fcc_us_no_bootcamp = fcc_top_countries[bool_no_bootcamp]

fcc_top_countries = fcc_top_countries.drop(fcc_us_no_bootcamp.index)
```

```python
## Removing the participants that were programming for 3 months or less when they completed the survey

bool_less_than_3_months = bool_us_outliers & (fcc_top_countries['MonthsProgramming'] <= 3)
fcc_us_less_than_3_months = fcc_top_countries[bool_less_than_3_months ]

fcc_top_countries = fcc_top_countries.drop(fcc_us_less_than_3_months.index)
```

```python
sns.boxplot(x = 'CountryLive', y = 'money_per_month', data = fcc_top_countries )
plt.xlabel('Country of Residence')
plt.ylabel('Money Per Month (USD)')
plt.title('Monthly Amount Spent on Learning', y = 1.02)
plt.xticks([0,1,2,3], ['US', 'UK', 'India','Canada'], rotation = 0)
#plt.ylim([0,80000])
plt.show()
```

#### We will also analyze the outlier for Canada. 

```python
## Analyzing the Canada outlier

bool_outlier_canada = (fcc_top_countries['CountryLive'] == 'Canada') & (fcc_top_countries['money_per_month'] > 3000)
canada_outlier = fcc_top_countries[bool_outlier_canada ]

canada_outlier[['MonthsProgramming', 'AttendedBootcamp']]
```

#### It seems that with Canada also, it is a similar case. The person was attending a bootcamp and paid a large sum at the start. Most likely he didn't pay in the subsequent months and hence can be removed as an outlier. 

```python
## Removing the Canada outlier

fcc_top_countries = fcc_top_countries.drop(canada_outlier.index)
```

```python
sns.boxplot(x = 'CountryLive', y = 'money_per_month', data = fcc_top_countries )
plt.xlabel('Country of Residence')
plt.ylabel('Money Per Month (USD)')
plt.title('Monthly Amount Spent on Learning', y = 1.02)
plt.xticks([0,1,2,3], ['US', 'UK', 'India','Canada'], rotation = 0)
#plt.ylim([0,80000])
plt.show()
```

#### Noting the updated mean spend values for the top 4 countries

```python
fcc_top_countries.groupby('CountryLive').mean()['money_per_month'].sort_values(ascending = False)
```

#### We can see that Canada has replaced India as the second highest spending location. But before making the final recommendation, we should also check for potential market size. In India, the per month spend is USD 65, which is more than the USD 59 cost for our subscription. If we get more consumers from India, we can recommend India as the second market over Canada. 

```python
fcc_top_countries['CountryLive'].value_counts(normalize = True).sort_values(ascending = False).head(4)*100
```

#### Upon observing the distribution (post removal of the outliers), we can see that India has twice as many coders in comparision to Canada. Hence we may recommend India as the second market. The top market remains the US, both in terms of consumer spending as well as Market size.
